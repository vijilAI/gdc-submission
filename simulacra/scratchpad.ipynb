```python
import sys
import os

# Add the project root directory to sys.path
sys.path.append(os.path.dirname(os.getcwd()))
# Or more explicitly
# sys.path.append('/home/smajumdar/vijil/poc_projects')

from dotenv import load_dotenv
load_dotenv("../.env")

from simulacra.agent_v2 import ConversationAgent, RedTeamerAgent, RedTeamingSession
```

```python
nora_bot = ConversationAgent(config_path="configs/nora.yaml")

# Add before the for loop
os.makedirs("sessions/nopersona", exist_ok=True)

# look into each file under /personas and create a bot for each
for persona_file in os.listdir("personas"):
    persona_name = persona_file[:-5]  # Remove the extension
    # red_bot = RedTeamerAgent(persona_path=f"personas/{persona_file}", config_path="configs/redteamer_nopersona.yaml")
    red_bot = RedTeamerAgent(persona_path=f"personas/{persona_file}", config_path="configs/redteamer_noocean.yaml")
    # red_bot = RedTeamerAgent(persona_path=f"personas/{persona_file}")
    
    print("red teaming using persona", persona_name)
    session = RedTeamingSession(
        sut_agent= nora_bot,
        redteamer_agent=red_bot,
    )
    await session.run_session(iters=2, max_turns=2)
    
    session.export_conversation_history(f"sessions/nopersona/{persona_name}.jsonl")
    # session.export_conversation_history(f"sessions/noocean/{persona_name}.jsonl")
    # session.export_conversation_history(f"sessions/deepseek_distil_qwen_14b/{persona_name}.jsonl")
```
